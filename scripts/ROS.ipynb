# ROS

## Εισαγωγή

Όμως δεν είναι ακριβώς OS όπως τα Windows.. <br>
Ουσιαστικά, είναι κάποιες βιβλιοθήκες με κώδικα που <<δένουν>> τα κομμάτια του ρομπότ. <br><br>

To ROS για να εγκατασταθεί χρειάζεται λογισμικό <abbr title="Το καλύτερο λογισμικό στον πλανήτη γη">Linux</abbr>. 
Δυστυχώς δεν μπορείται στα <abbr title="Το χειρότερο λογισμικό στο πλανήτη γη">Windows</abbr><font size='2'><sup>[1]</sup></font> ... <br>

  1. Πλοηγηθήται στην ιστοσελίδα του και φτιάξτε λογαριασμό <br>
  2. Συνδεθείται και πατήστε το MyRosProjects όπως φαίνεται αριστερά στην εικόνα <br>
  3. Φτιάξτε νέο ROSject με το '+' στην εικόνα <br>
  4. Δώστε ότι όνομα θέλετε και επιλέξτε για <span style="color:red">ROS Distro </span> το <span style="color:red">ROS Melodic </span>. Με 'Dist' εννοεί 'Distribution' και είναι οι [εκδόσεις](http://wiki.ros.org/Distributions "Κάθε 1-2 χρόνια βγάζουν και από μια") του ROS. Επιλέγουμε το Melodic Distro επειδή αυτό χρησιμοποιούμε στην ομάδα αλλά μπορείτε και αυθαίρετα. <br>

## Βασικές έννοιες

### Catkin_make

Στην αρχή κάθε project ακολουθούμε μια διαδικασία:
1. Φτιάχνουμε ένα φάκελο (<abbr title="Το παίρνει το ποτάμι: με το mkdir">όπως μάθατε</abbr>) που θα είναι το  <span style="color:fuchsia">workspace</span>. Για να είναι τα πράγματα πιο ευανάγνωστα η ονομασία του φακέλου είναι '<span style="color:aqua">κάτι</span><span style="color:fuchsia">_ws</span>' όπου '<span style="color:aqua">κάτι</span>' βάζεται ότι θέλεται και '<span style="color:fuchsia">_ws</span>' σημαίνει <span style="color:fuchsia">workspace</span>. Π.χ. εγώ το ονόμασα '<span style="color:aqua">myws</span><span style="color:fuchsia">_ws</span>'
<br> Παρατηρήστε ότι και το code editor (από πίσω) δείχνει τους φακέλους.
2. Έπειτα <abbr title="Το παίρνει το ποτάμι: με το cd">μπαίνουμε</abbr> μέσα σε αυτόν τον φάκελο που φτιάξαμε και φτιάχνουμε νέο φάκελο, τον '</span><span style="color:fuchsia">src</span>'. Το όνομα βγαίνει από το source ή πιο συγκεκριμένα το source code. Δηλαδή εκεί θα μπαίνει ο πηγαίος μας κώδικας!
3. Τέλος <abbr title="Το παίρνει το ποτάμι: με το cd..">βγαίνουμε</abbr> από τον φάκελο και πληκτρολογούμε catkin_make ! <br>
Το [catkin](http://wiki.ros.org/catkin "Στο the constrct είναι ήδη κατεβασμένο αλλά κανονικά κατεβαίνει με το ROS") είναι μια βιβλιοθήκη που βοηθάει στην δημιουργία workspace και catkin_**make** είναι η εντολή που το φτιάχνει!



αυτή η διαδικασία είναι τιφλοσούρτι

### Catkin_create_pkg

Τέλεια! Τώρα που έχουμε το workspace μας θα βάλουμε μέσα κώδικα. <br>
Για να είμαστε οργανωμένοι ο κώδικας κατανέμεται σε **πακέτα** (**packages**). Π.χ. μπορεί να έχεις ένα πακέτο (ουσιαστικά έναν φάκελο) που είναι αποκλειστικά για τις ρόδες του ρομπότ. Όλος ο κώδικας για τις ρόδες θα μπει εκεί. <br>
Τα πακέτα μπαίνουν μέσα στο src. Μπείτε στο src και πληκτρολογίστε την παρακάτω εντολή:

```python
catkin_create_pkg beginner_tutorials std_msgs rospy roscpp
```

Η εντολή "**catkin_create_pkg**" είναι αυτή που φτιάχνει το πακέτο επονόματι "beginner_tutorials" και έπειτα λέει ότι το πακέτο εξαρτάται από κάποια dependencies, τα "std_msgs", "rospy", "roscpp". <br>
Μην σαν ανησυχούν τα dependencies, αρκεί να ξέρετε ότι το **rospy** & **roscpp** είναι οι βιβλιοθήκες ROS για τις γλώσσες Python & C++ αντίστοιχα. <br> <br>

### Publisher-Subscriber (ή αλλιώς Talker-Listener)

Θα κατεβάσουμε έναν έτοιμο workspace για να δούμε την βασική έννοια των **publisher/subscriber**. <br>
Πληκτρολογήστε τις παρακάτω εντολές με την σειρά:

```python
cd ~
git clone https://github.com/huegli/ros-talker-listener
cd ros-talker-listener/my_ws
catkin_make
cd ros-talker-listener/my_ws/src/resin-talker-listener/
ls
```

Μπήκατε στο workspace my_ws και το πακέτο resin-talker-listener που κατεβάσαμε από το ίντερνετ. <br>
Βλέπετε με το ls ότι περιέχει τους φακέλους scripts και launch. Ας τους εξηγήσουμε:
* **scripts**: Εκεί μπαίνει ο κώδικας Python, C++ κλπ 
* **launch**: Εκεί μπαίνει ο κώδικας που θα <<λανσάρει>> τα scripts. (π.χ. μπορούμε από ένα launch αρχείο να λανσάρουμε 10 scrits με ότι σειρά και παραμέτρους θέλουμε)



#### Roslaunch

Πως το κάνουμε αυτό? Με την εντολή roslaunch! Εκτελέστε τα παρακάτω:

```python
cd ~/ros-talker-listener/my_ws/
source devel/setup.bash
roslaunch resin-talker-listener default.launch
```

Στην τελευταία γραμμή, η εντολή "**roslaunch**" λανσάρει το "**default.launch**" αρχείο από το πακέτο "**ros-talker-listener**". <br>
Μέσα του το launch αρχείο λανσάρει το talker.py αρχείο (κώδικας Python) που βρίσκεται στον φάκελο scripts <br><br>
<span style="color:red">ΠΡΟΣΟΧΗ!!!</span>: Εάν δεν εκτελέσετε την γραμμή "source devel/setup.bash" το ROS δεν θα ξέρει που κοιτάει για πακέτα! <br>
Όταν την εκτελείτε του λέτε "Ε ROS! Εδώ είναι το workspace μου, εδώ ψάχνε τα πακέτα!"

#### Rosrun

Τώρα τρέχει ο talker, δηλαδή αυτός που μιλάει αλλά κανείς δεν τον ακούει  :'( <br>
Ας βάλουμε κάποιον να τον ακούει. <br>
Ανόιξτε νέα καρτέλα web shell και εκτελέστε τις παρακάτω εντολές:

```python
cd ros-talker-listener/my_ws
source devel/setup.bash
rosrun resin-talker-listener listener.py
```

Στην τελευταία γραμμή, η εντολή "**rosrun**" εκτελεί το "**listener.py**" αρχείο από το πακέτο "**resin-talker-listener**". <br>
Θα μπορούσαμε να το είχαμε λανσάρει μέσω ενός άλλου ή και του ίδιου launch αρχείου αλλά εναλλακτικά η εντολή rosrun μας επιτρέπει να το <<τρέξουμε>> (run) απευθείας.

Βλέπετε ότι κάθε στιγμή ο listener ακούει από τον talker το "Hello World!" και το εκτυπώνει στην οθόνη! 
 <br>
Με αυτόν τον τρόπο επικοινωνούν τα διάφορα εξαρτήματα του ρομπότ. Το καθένα <<μιλάει>> σε άλλα και <<ακούει>> από άλλα <br><br>
Εάν πατήσετε "rqt_graph" σε μια νέα καρτέλα θα σας εμφανιστεί το εξής: 



Εδώ φαίνονται οι κόμβοι (nodes) με κύκλους και τα θέματα (topics) με ορθογώνια παραλληλόγραμμα. <br>
Κάθε node είναι ένας κώδικας .py (Python) και κάθε topic είναι η <<συζήτηση>> τους, δηλαδή το κανάλι ροής με το οποίο επικοινωνούν. <br>
Ο κόμβος talker μιλάει στον listener κι αυτός ακούει.
